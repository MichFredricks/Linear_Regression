Implementation of linear regression in python using the data and ideas taught in Andrew Ng's Machine Learning Coursera course (originally taught in MATLAB)

```python
import numpy as np
import matplotlib.pyplot as plt
```

Load Data (Profits of food trucks as a function of populations of the cities they are in)

```python
X = []
Y = []
data_file = open("ex1data1.txt","r")
for dat in data_file:
    d = dat.split(",")
    if d[:-1]:        
        X.append(float(d[0]))
        Y.append(float(d[-1].strip("\n")))
```

Visualize Data

```python
plt.figure(figsize=(20,10))
plt.scatter(X,Y)
plt.show()
```

Add vector of ones to X to enable bias in linear model

```python
X = np.column_stack((np.ones(len(X)),np.array(X)))
Y = np.array(Y)
```

Define linear model, cost functions, and gradient descent functions

```python
def linear_model(X,Theta):
    return np.inner(X,Theta)
    
def cost(X,Theta,Y):
    return np.sum(np.square(linear_model(X,theta)-Y))/(2*len(Y))

def gradient_descent(X,Theta,Y,alpha):
    return Theta - alpha*((np.transpose(X)*(np.inner(X,Theta)-Y)).sum(axis=1))/len(Y)
```

Run Gradient Descent

```python
theta = [-1,2]
print("Initial cost:",cost(X,theta,Y))
for i in range(1500):
    theta = gradient_descent(X,theta,Y,0.01)
print("Final cost:",cost(X,theta,Y))
```

Plot data with linear model

```python
plt.figure(figsize=(20,10))
plt.scatter([dat[1] for dat in X],Y)
x = ([min([dat[1] for dat in X]),max([dat[1] for dat in X])])
y = [theta[0]+theta[1]*dat for dat in x]
plt.plot(x,y,'k-')
plt.show()
```
